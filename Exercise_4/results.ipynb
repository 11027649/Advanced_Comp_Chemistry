```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

df = pd.read_csv("asm_results.txt", delimiter="\s+")
print(df.head())

print(df["Elstat"])

plt.plot(df["bondlength"], df["StrainTotal"], label="dE strain")
plt.plot(df["bondlength"], df["EnergyTotal"], label="dE")
plt.plot(df["bondlength"], df["Int"], label="dE intern")

plt.ylabel("Energy")
plt.xlabel("Reaction coordinate (bondlength)")
plt.legend()
```
