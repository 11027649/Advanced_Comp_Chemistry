```python
# make GO script for halogens
atoms = ['F ', 'Br', 'Cl']

for atom1 in atoms:
    for atom2 in atoms:
        for atom3 in atoms:
            print(atom1, atom2, atom3)

            with open('GO_empty.txt', 'r') as runfile:
                script = runfile.readlines()
                
                with open('BrClFcoordinates.xyz', 'r') as coordinates:
                    lines = coordinates.readlines()
                    
                    # first line is middle atom
                    lines[0] = atom2 + lines[0][3:]
                    
                    lines[1] = atom1 + lines[1][3:]
                    
                    # ion
                    lines[2] = atom3 + lines[2][3:]

            filename = 'halogen/go' + atom1.strip() + atom2.strip() + atom3.strip()
            
            with open(filename, 'w') as f:
                for line in script:
                    
                    if line.startswith('ATOM'):
                        f.writelines([line])
                        
                        f.writelines([lines[0], lines[1], lines[2]])
                    else:
                        f.writelines([line])
                
```

```python
# make GO script for hydrogens
atoms = ['F ', 'Br', 'Cl']

for atom1 in atoms:
    for atom2 in atoms:
            print(atom1, 'H', atom2)

            with open('GO_empty.txt', 'r') as runfile:
                script = runfile.readlines()
                
                with open('ClHBrcoordinates.xyz', 'r') as coordinates:
                    lines = coordinates.readlines()
                    
                    # first line is middle atom
                    lines[0] = atom1 + lines[0][3:]
                    
                    lines[2] = atom2 + lines[2][3:]

            filename = 'hydrogen/go' + atom1.strip() + 'H' + atom2.strip()
            
            with open(filename, 'w') as f:
                for line in script:
                    
                    if line.startswith('ATOM'):
                        f.writelines([line])
                        
                        f.writelines([lines[0], lines[1], lines[2]])
                    else:
                        f.writelines([line])
                
```

```python
# make Single Point script for hydrogens
import os 

for directory in os.listdir('hydrogen'):
    
    # get coordinates from go.out file
    with open('hydrogen/' + directory + '/go' + directory + '.out', 'r') as coordinates_file:
        lines = coordinates_file.readlines()
        
        for i, line in enumerate(lines):
            if 'Coordinates in Geometry Cycle' in line:
                atom1 = lines[i + 2]
                atom2 = lines[i + 3]
                atom3 = lines[i + 4]

                frag1 = atom1[6] + atom1[7].strip() + atom2[6] + atom2[7].strip()
                frag2 = atom3[6] + atom3[7].strip()
                print(frag1, frag2)
    
    with open('SP_empty.txt', 'r') as empty_script:
        lines = empty_script.readlines()
        
    # make script for first fragment
    with open('hydrogen/' + directory + '/sp_frag1_' + frag1, 'w') as f:
        for line in lines:
            if line.startswith('ATOM'):
                f.writelines([line])

                f.writelines([atom1, atom2])
            else:
                f.writelines([line])
                
    # make script for first fragment
    with open('hydrogen/' + directory + '/sp_frag2_' + frag2, 'w') as f:
        for line in lines:
            if line.startswith('ATOM'):
                f.writelines([line])

                f.writelines([atom3])
            else:
                f.writelines([line])
            
        
    
```
